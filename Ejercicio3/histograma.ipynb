```python
import pandas as pd
import matplotlib.pyplot as plt
import random
import numpy as np
```

```python
datos = []
x = []

for y in range(256):
    x.append(y+1)

with open("r.txt") as archivo:
    for linea in archivo:
        datos.append(int(linea))


plt.hist(datos, bins = 255, color = "red", rwidth=1)
plt.title("Frecuencia de intensidad del color Rojo")
plt.xlabel("Intensidad")
plt.ylabel("Numero de Pixeles")
plt.show()

```
